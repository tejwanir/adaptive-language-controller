```python
import whisper
import whisper_timestamped
import IPython.display as ipd

model = whisper.load_model("medium.en")
```

```python
audio = whisper.load_audio("./IMG_4453.MOV")
```

```python
output = whisper_timestamped.transcribe(model, audio, language="en")
for segment in output["segments"]:
    time_start = segment["start"]
    time_end = segment["end"]
    t = lambda t: f"{f'{t:.2f}'.rjust(5, '0')}s"
    text = segment["text"]
    print(f"{t(time_start)} - {t(time_end)}: {text}")
    start_idx = int(time_start * whisper.audio.SAMPLE_RATE)
    end_idx = int(time_end * whisper.audio.SAMPLE_RATE + 0.5)
    display(ipd.Audio(audio[start_idx:end_idx], rate=whisper.audio.SAMPLE_RATE))
```
